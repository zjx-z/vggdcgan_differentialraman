```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.linalg import svd

# 清除环境并加载数据
# a = np.loadtxt('shujvzhengli.txt')  # 如果是文本文件
a = pd.read_excel("zhuanzhiduiyingfenxi.xlsx").values
#a=pd.read_excel("vggduiyingfenxi.xlsx").values
print(a.size)
a1=np.isinf(a)
a2=np.isnan(a)
for i in range(39):
    for j in range(71):
        if a1[i][j]==True or a2[i][j]==True:
            print(a[i][j])
        if a[i][j]==0:
            a[i][j]=1#vgg就要0.00001了
# filtered_a = a[~np.isinf(a) & ~np.isnan(a)]
# print(filtered_a)
# 计算总和、概率矩阵、行和列的和
T = np.sum(a)
P = a / T
r = np.sum(P, axis=1)
c = np.sum(P, axis=0)
print(c.size)

# 计算行和列的分布
Row_profile = a / np.tile(np.sum(a, axis=1), (a.shape[1], 1)).T
B = (P - np.outer(r, c)) / np.sqrt(np.outer(r, c))

# 奇异值分解
u, s, vh = svd(B, full_matrices=False)
w1 = np.sign(np.sum(vh, axis=1))
w2 = np.sign(np.sum(u, axis=0))
vb = vh * w1[:, np.newaxis]
ub = u * w2[np.newaxis, :]

# 计算 lambda 和 ksi2square
lambda_ = np.diag(s) ** 2
ksi2square = T * lambda_
T_ksi2square = np.sum(ksi2square)
con_rate = lambda_ / np.sum(lambda_)
cum_rate = np.cumsum(con_rate)

# 计算 beta 和 G
beta = np.diag(1.0 / np.sqrt(r)) @ ub
G = beta @ np.diag(s)
print(vb.shape)
# 计算 alpha 和 F
alpha = np.diag(1.0 / np.sqrt(c)) @ vb.T
F = alpha @ np.diag(s)

# # 可视化部分
# # 这里假设 yb 和 ch 是已经定义好的行标签和列标签
# #yb = np.array([""])  # 根据实际情况定义
# #ch = np.array([...])  # 根据实际情况定义
# import numpy as np

# # 生成 ch 数组，包含从 1 到 67 的整数
ch = np.arange(1, 72)  # np.arange 的结束值是不包含的，所以使用 68

# # 生成 yb 数组，包含从 1 到 39 的整数
yb = np.arange(1, 40)  # 同上，使用 40 来包含 39

# # 将整数数组转换为字符串数组，如果需要的话
# ch_str = ch.astype(str)
# yb_str = yb.astype(str)

plt.figure()
plt.plot(F[:, 0], F[:, 1], '.', color='k', linewidth=2)
# plt.text(F[:, 0] + 0.2, F[:, 1], ch)
# plt.xlabel('dim1')
# plt.ylabel('dim2')
# plt.show()
L=F[:,0:2]
x=[]
y=[]
z=[]
for i in L:
    x.append(i[0])
    y.append(i[1])
    #z.append(i[2])
```
