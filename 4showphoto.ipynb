```python
#本代码功能为展示某文件夹下所有图片，在聚类任务中可以直观观察哪些被错误归类

from torchvision.datasets import ImageFolder
import torchvision 
import torch
from torchvision import transforms, datasets
import torch.nn as nn
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.manifold import TSNE
transforms = transforms.Compose([
    transforms.Resize(256),    # 将图片短边缩放至256，长宽比保持不变：
    #transforms.CenterCrop(224),   #将图片从中心切剪成3*224*224大小的图片
    transforms.ToTensor()          #把图片进行归一化，并把数据转换成Tensor类型
]) 
path = r'E:/milkbottle\show'#在里面只能由一个文件夹，再里面才放图
 
data_train = datasets.ImageFolder(path, transform=transforms)
#datatolist=data_train.tolist()
 
data_loader = DataLoader(data_train, batch_size=len(data_train), shuffle=True)
 
#X=[]
#for i, data in enumerate(data_loader): 
#    images, labels = data
#    X.append(datatolist[i])
#tsne = TSNE(n_components=2)
#X_tsne = tsne.fit_transform(X)


nrow=12#每行十个图
plt.figure(figsize=(10, 8))  # 可以根据需要调整这个值，这里设置为宽10英寸，高8英寸
for i, data in enumerate(data_loader):
    images, labels = data
    #print(images)
    # 打印数据集中的图片
    img = torchvision.utils.make_grid(images,nrow=nrow).numpy()
    plt.imshow(np.transpose(img, (1, 2, 0)))
    plt.show()
 
```
